```python
import numpy as np
import pandas as pd

df=pd.read_csv('property_data.csv') # to load .csv file to the data frame
df
```

```python
df.head(12) # head() will return first 5 rows of the dataframe by default
```

```python
df.dtypes # returns type of every columns
```

```python
# some columns have values that defer from 'dtype' result, so we have to do data cleaning
# 'NAN' means that data has no value
```

```python
droped_df=df.dropna() # this command removes all rows with 'NAN' value
droped_df
```

```python
# this means most of the rows posses 'NAN' value
```

```python
droped_df=df.dropna(axis=1) # this command removes all columns with 'NAN' value
droped_df
```

```python
droped_df=df.dropna(thresh=int(df.shape[0] * .9), axis=1) # only those columns that have less than 90% clean data will be removed 
droped_df
```

```python
df
```

```python
# to identify whether this column has 'NAN' values
df['ST_NUM'].isnull()
```

```python
# this code returns all the columns that have 'NAN' value

column_name=df.columns
NAN_columns=list()     


for col in column_name:
    if (True in np.array(df[col].isnull())):
        NAN_columns.append(col)

        
NAN_columns  # list of columns contains NAN
```

```python
df['ST_NUM']=df['ST_NUM'].fillna(10)   # fill the NAN values in ST_NUM column with 10
df['ST_NUM']
```

```python
df['PID']=df['PID'].fillna(df['PID'].mean()) # fill NAN value of PID column with mean() of PID 
df['PID'] 
```

```python
# now we can clean variant values of every column, to do so we first replace them with NAN
# this for loop replace every variant value of column with NAN

for i in range(len(df['NUM_BATH'])):
    try:
        int(df['NUM_BATH'][i]) 
    except ValueError:
        df.loc[i, 'NUM_BATH'] = np.nan # replace with NAN

df.head()
```

```python
# we can replace variant values of each column with NAN

missing_value=['n/a', 'na', '--']
new_df=pd.read_csv('property_data.csv', na_values=missing_value)
new_df
```
