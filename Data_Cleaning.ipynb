```python
droped_df=df.dropna(thresh=int(df.shape[0] * .9), axis=1) # only those columns that have less than 90% clean data will be removed 
droped_df
```

```python
# to identify whether this column has 'NAN' values
df['ST_NUM'].isnull()
```

```python
# this code returns all the columns that have 'NAN' value

column_name=df.columns
NAN_columns=list()     


for col in column_name:
    if (True in np.array(df[col].isnull())):
        NAN_columns.append(col)

        
NAN_columns  # list of columns contains NAN
```

```python
df['ST_NUM']=df['ST_NUM'].fillna(10)   # fill the NAN values in ST_NUM column with 10
df['ST_NUM']
```

```python
df['PID']=df['PID'].fillna(df['PID'].mean()) # fill NAN value of PID column with mean() of PID 
df['PID'] 
```

```python
# now we can clean variant values of every column, to do so we first replace them with NAN
# this for loop replace every variant value of column with NAN

for i in range(len(df['NUM_BATH'])):
    try:
        int(df['NUM_BATH'][i]) 
    except ValueError:
        df.loc[i, 'NUM_BATH'] = np.nan # replace with NAN

df.head()
```

# Pre Processing using SKLEARN

```python
from sklearn import preprocessing
import numpy as np


rand_array=np.random.randint(-100, 100, 10) # random int array
print(rand_array)

ary_str = ['foo', 'bar', 'baz', 'x', 'y', 'z'] # create an array for preprocessing
```

```python
# to encode cathegorical data (change to numerical)

l_encoder=preprocessing.LabelEncoder()
l_encoder.fit(ary_str)
print(l_encoder.transform(['foo']))
print(l_encoder.transform(['baz']))
print(l_encoder.transform(['bar']))
```

```python
list(l_encoder.classes_)  # create mapping
```

```python
# create a dictionary to store mapping

dict(zip(l_encoder.classes_, l_encoder.transform(l_encoder.classes_)))
```

```python
# using DataFrames instead

import pandas as pd
l_encoder=preprocessing.LabelEncoder()
df = pd.DataFrame(data = {'col1': ['foo','bar','foo','bar'],
                          'col2': ['x', 'y', 'x', 'z'], 
                          'col3': [1, 2, 3, 4]})
df.apply(l_encoder.fit_transform)
```

```python
# ordinal encoder

o_encoder=preprocessing.OrdinalEncoder()
ary_2d=[['foo', 'bar', 'baz'],['x', 'y', 'z']]
o_encoder.fit(ary_2d) # fit the values
o_encoder.transform([['foo', 'y', 'z']])
```

```python
# onehot # Hot encoder

hot_encoder=preprocessing.OneHotEncoder(handle_unknown='ignore')
hot_encoder.fit(ary_2d)
hot_encoder.categories_
```

```python
hot_encoder.transform([['foo', 'foo', 'baz'], ['y', 'y', 'x']]).toarray()
```
